```python
def time_slot(hour):
    if 0 <= hour < 6: return 'Late Night'
    elif 6 <= hour < 12: return 'Morning'
    elif 12 <= hour < 18: return 'Afternoon'
    else: return 'Evening'
df['TimeSlot'] = df['PurchaseHour'].apply(time_slot)
```

```python
def categorize_browsing(time):
    if time < 2:
        return 'Quick Browser'
    elif 2 <= time < 5:
        return 'Average Browser'
    elif 5 <= time < 10:
        return 'In-depth Browser'
    else:
        return 'Long Researcher'

df['BrowsingCategory'] = df['BrowsingTime'].apply(categorize_browsing)
```

```python
df['BrowsingCategory']
```

```python
def assign_age_group(age):
    if pd.isnull(age):
        return 'Unknown'
    elif age < 18:
        return 'Teen'
    elif 18 <= age < 25:
        return 'Youth'
    elif 25 <= age < 45:
        return 'Adult'
    elif 45 <= age < 60:
        return 'Middle-Aged'
    else:
        return 'Senior'
```

```python
df['AgeGroup'] = df['Age'].apply(assign_age_group)
```

## Clustering Techiniques

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


features = df.groupby('CustomerID').agg({
    'Age': 'mean',
    'Total_Sales': 'sum',
    'Quantity': 'sum',
    'BrowsingTime': 'mean',
    'Device': lambda x: x.mode()[0],  # most used device
    'City': lambda x: x.mode()[0]     # most common city
}).reset_index()

# Encode categorical features
features = pd.get_dummies(features, columns=['Device', 'City'], drop_first=True)

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features.drop('CustomerID', axis=1))
```

```python
from sklearn.cluster import KMeans

# Find optimal number of clusters using Elbow Method
wcss = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(2, 10), wcss, marker='o')
plt.title('Elbow Method - Optimal Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

```python
# Fit final model
kmeans = KMeans(n_clusters=4, random_state=42)
features['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
cluster_summary = features.groupby('Cluster').mean()
display(cluster_summary)
```

| ------- | --------------------- | ------------------------ | ------------------------------------------------------------------------------------- |
| 0       | Bargain Shoppers      | Value Seekers          | Emphasizes their focus on low spend but active browsing—looking for deals.            |
| 1       | Loyal High-Spenders   | Premium Loyalists      | Highlights both their high spending and brand loyalty.                                |
| 2       | Impulse Buyers        | Spontaneous Spenders   | More relatable and polished wording for impulse behavior.                             |
| 3       | Window Shoppers       | Curious Browsers       | Indicates high browsing with low purchase intent, more modern than "window shoppers". |

## RFM Analysis 

```python
## step 1:
df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'])

# Reference date: usually one day after last transaction
latest_date = df['PurchaseDate'].max() + pd.Timedelta(days=1)
```

```python
##  step 2 :Calculate RFM Metrics
rfm = df.groupby('CustomerID').agg({
    'PurchaseDate': lambda x: (latest_date - x.max()).days,  # Recency
    'CustomerID': 'count',                                    # Frequency
    'Total_Sales': 'sum'                                      # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm = rfm.reset_index()
```

```python
## Step 3: Rank Each Metric

rfm['R_rank'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])  # Lower recency = better
rfm['F_rank'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])
rfm['M_rank'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])
```

```python
## Step 4: Create RFM Score

rfm['RFM_Score'] = rfm['R_rank'].astype(str) + rfm['F_rank'].astype(str) + rfm['M_rank'].astype(str)
```

```python
##Step 5: Segment Based on RFM Score

def segment_customer(df):
    score = df['RFM_Score']
    if score in ['444', '434', '344', '443']:
        return 'Champions'
    elif score in ['344', '343', '334', '333']:
        return 'Loyal Customers'
    elif score.startswith('1'):
        return 'At Risk'
    elif score.endswith('1'):
        return 'Low Spenders'
    else:
        return 'Others'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```

```python
## Step 6: Analyze & Visualize Segments
## Count per Segment
rfm['Segment'].value_counts().plot(kind='bar', color='teal', figsize=(8,4), title="Customer Segments")
```

```python
## Monetary Value per Segment

rfm.groupby('Segment')['Monetary'].mean().sort_values().plot(kind='barh', title="Avg Spend by Segment")
```

```python
## Step-by-Step: Heatmap of RFM Segment vs City

# From earlier RFM processing
# rfm has: CustomerID, Segment

# Merge it with city information
rfm = rfm.merge(df[['CustomerID', 'City']], on='CustomerID', how='left')
```

```python
##  Create a Pivot Table for the Heatmap

heatmap_data = rfm.pivot_table(index='Segment', 
                                columns='City', 
                                values='CustomerID', 
                                aggfunc='count').fillna(0)
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(heatmap_data, 
            annot=True, 
            fmt=".0f", 
            cmap="YlGnBu", 
            linewidths=.5)

plt.title("RFM Segments vs Cities - Customer Count Heatmap", fontsize=16)
plt.xlabel("City")
plt.ylabel("RFM Segment")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
